```python
# ============================================================
# CNN BASELINE EXPERIMENTS
# Optimize edilmemiş CNN modelleri
# - Data augmentation yok
# - Differential learning rate yok
# - Katman katman fine-tuning yok
# - Sadece son classifier/fc eğitilir
# - Pretrained ImageNet ağırlıkları kullanılır
# - 5-Fold + best fold final test
# ============================================================


# ============================================================
# 1. KÜTÜPHANELER
# ============================================================

import os
import gc
import copy
import shutil
import numpy as np
import pandas as pd

import torch
import torch.nn as nn
import torch.optim as optim

from torch.utils.data import DataLoader, Subset
from torchvision import datasets, transforms, models

from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
    confusion_matrix,
    roc_curve
)

import matplotlib.pyplot as plt
import seaborn as sns

from google.colab import drive
from IPython.display import display


# ============================================================
# 2. DRIVE BAĞLAMA VE DATASET AÇMA
# ============================================================

drive.mount('/content/drive')

rar_source_path = "/content/drive/MyDrive/data_set_zip/breast_mri_dataset.rar"
rar_target_path = "/content/breast_mri_dataset.rar"

shutil.copy(rar_source_path, rar_target_path)

# Aynı dosyalar varsa üzerine yazsın.
os.system(f'unrar x -o+ "{rar_target_path}" /content/')


# ============================================================
# 3. GENEL AYARLAR
# ============================================================

DATA_DIR = "/content/breast_mri_dataset"

BATCH_SIZE = 32
NUM_EPOCHS = 10
NUM_CLASSES = 2

BASE_LR = 1e-3
N_SPLITS = 5
RANDOM_STATE = 42

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Cihaz: {device}")

# Hepsini çalıştırmak için:
RUN_MODELS = [
    "resnet34",
    "densenet121",
    "efficientnet_b0",
    "mobilenet_v2",
    "inception_v3"
]

# Sadece tek model çalıştırmak istersen örnek:
# RUN_MODELS = ["resnet34"]
# RUN_MODELS = ["densenet121"]
# RUN_MODELS = ["efficientnet_b0"]
# RUN_MODELS = ["mobilenet_v2"]
# RUN_MODELS = ["inception_v3"]


# ============================================================
# 4. ÇIKTI KLASÖRLERİ
# ============================================================

OUTPUT_ROOT = "/content/drive/MyDrive/breast_mri_cnn_baseline_outputs"

os.makedirs(OUTPUT_ROOT, exist_ok=True)

combined_summary_rows = []


# ============================================================
# 5. SEED AYARI
# ============================================================

def set_seed(seed=42):
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)


set_seed(RANDOM_STATE)


# ============================================================
# 6. MODEL BİLGİLERİ
# ============================================================

MODEL_CONFIGS = {
    "resnet34": {
        "display_name": "ResNet34 Baseline",
        "img_size": 224,
        "excel_name": "resnet34_baseline_results.xlsx"
    },
    "densenet121": {
        "display_name": "DenseNet121 Baseline",
        "img_size": 224,
        "excel_name": "densenet121_baseline_results.xlsx"
    },
    "efficientnet_b0": {
        "display_name": "EfficientNet-B0 Baseline",
        "img_size": 224,
        "excel_name": "efficientnet_b0_baseline_results.xlsx"
    },
    "mobilenet_v2": {
        "display_name": "MobileNetV2 Baseline",
        "img_size": 224,
        "excel_name": "mobilenet_v2_baseline_results.xlsx"
    },
    "inception_v3": {
        "display_name": "InceptionV3 Baseline",
        "img_size": 299,
        "excel_name": "inception_v3_baseline_results.xlsx"
    }
}


# ============================================================
# 7. TRANSFORM FONKSİYONU
# ============================================================

def get_baseline_transforms(img_size):
    mean = [0.485, 0.456, 0.406]
    std = [0.229, 0.224, 0.225]

    baseline_transform = transforms.Compose([
        transforms.Resize((img_size, img_size)),
        transforms.ToTensor(),
        transforms.Normalize(mean, std)
    ])

    return baseline_transform


# ============================================================
# 8. BASELINE MODEL OLUŞTURMA
# ============================================================

def build_baseline_model(model_name):
    """
    Baseline mantığı:
    - ImageNet pretrained model alınır.
    - Tüm backbone dondurulur.
    - Sadece son classifier / fc katmanı değiştirilir ve eğitilir.
    """

    if model_name == "resnet34":
        model = models.resnet34(
            weights=models.ResNet34_Weights.IMAGENET1K_V1
        )

        for param in model.parameters():
            param.requires_grad = False

        num_ftrs = model.fc.in_features
        model.fc = nn.Linear(num_ftrs, NUM_CLASSES)

    elif model_name == "densenet121":
        model = models.densenet121(
            weights=models.DenseNet121_Weights.IMAGENET1K_V1
        )

        for param in model.parameters():
            param.requires_grad = False

        num_ftrs = model.classifier.in_features
        model.classifier = nn.Linear(num_ftrs, NUM_CLASSES)

    elif model_name == "efficientnet_b0":
        model = models.efficientnet_b0(
            weights=models.EfficientNet_B0_Weights.IMAGENET1K_V1
        )

        for param in model.parameters():
            param.requires_grad = False

        num_ftrs = model.classifier[1].in_features
        model.classifier[1] = nn.Linear(num_ftrs, NUM_CLASSES)

    elif model_name == "mobilenet_v2":
        model = models.mobilenet_v2(
            weights=models.MobileNet_V2_Weights.IMAGENET1K_V1
        )

        for param in model.parameters():
            param.requires_grad = False

        num_ftrs = model.classifier[1].in_features
        model.classifier = nn.Sequential(
            nn.Dropout(p=0.2),
            nn.Linear(num_ftrs, NUM_CLASSES)
        )

    elif model_name == "inception_v3":
        model = models.inception_v3(
            weights=models.Inception_V3_Weights.IMAGENET1K_V1,
            aux_logits=True
        )

        for param in model.parameters():
            param.requires_grad = False

        num_ftrs = model.fc.in_features
        model.fc = nn.Linear(num_ftrs, NUM_CLASSES)

        if model.AuxLogits is not None:
            aux_ftrs = model.AuxLogits.fc.in_features
            model.AuxLogits.fc = nn.Linear(aux_ftrs, NUM_CLASSES)

    else:
        raise ValueError(f"Bilinmeyen model adı: {model_name}")

    return model.to(device)


# ============================================================
# 9. OPTIMIZER
# ============================================================

def get_baseline_optimizer(model):
    """
    Sadece requires_grad=True olan parametreler eğitilir.
    Baseline için tek learning rate kullanılır.
    """

    trainable_params = filter(
        lambda p: p.requires_grad,
        model.parameters()
    )

    optimizer = optim.Adam(
        trainable_params,
        lr=BASE_LR
    )

    return optimizer


# ============================================================
# 10. INCEPTION OUTPUT YARDIMCI FONKSİYONU
# ============================================================

def get_main_output(outputs):
    if hasattr(outputs, "logits"):
        return outputs.logits

    if isinstance(outputs, tuple):
        return outputs[0]

    return outputs


# ============================================================
# 11. MODEL AĞIRLIKLARINI CPU'YA KOPYALAMA
# ============================================================

def clone_state_dict_to_cpu(model):
    return {
        key: value.detach().cpu().clone()
        for key, value in model.state_dict().items()
    }


# ============================================================
# 12. EĞİTİM FONKSİYONU
# ============================================================

def train_one_epoch(model, loader, optimizer, criterion):
    model.train()

    running_loss = 0.0
    correct = 0

    for images, labels in loader:
        images = images.to(device)
        labels = labels.to(device)

        optimizer.zero_grad()

        outputs = model(images)

        # InceptionV3 train modunda aux output üretir.
        if hasattr(outputs, "logits"):
            main_outputs = outputs.logits
            aux_outputs = outputs.aux_logits

            loss_main = criterion(main_outputs, labels)

            if aux_outputs is not None:
                loss_aux = criterion(aux_outputs, labels)
                loss = loss_main + 0.4 * loss_aux
            else:
                loss = loss_main

        elif isinstance(outputs, tuple):
            main_outputs = outputs[0]
            aux_outputs = outputs[1] if len(outputs) > 1 else None

            loss_main = criterion(main_outputs, labels)

            if aux_outputs is not None:
                loss_aux = criterion(aux_outputs, labels)
                loss = loss_main + 0.4 * loss_aux
            else:
                loss = loss_main

        else:
            main_outputs = outputs
            loss = criterion(main_outputs, labels)

        loss.backward()
        optimizer.step()

        running_loss += loss.item() * images.size(0)

        _, preds = torch.max(main_outputs, 1)
        correct += torch.sum(preds == labels.data).item()

    epoch_loss = running_loss / len(loader.dataset)
    epoch_acc = correct / len(loader.dataset)

    return epoch_loss, epoch_acc


# ============================================================
# 13. VALIDATION FONKSİYONU
# ============================================================

def evaluate_one_epoch(model, loader, criterion):
    model.eval()

    running_loss = 0.0
    correct = 0

    with torch.no_grad():
        for images, labels in loader:
            images = images.to(device)
            labels = labels.to(device)

            outputs = model(images)
            outputs = get_main_output(outputs)

            loss = criterion(outputs, labels)

            running_loss += loss.item() * images.size(0)

            _, preds = torch.max(outputs, 1)
            correct += torch.sum(preds == labels.data).item()

    epoch_loss = running_loss / len(loader.dataset)
    epoch_acc = correct / len(loader.dataset)

    return epoch_loss, epoch_acc


# ============================================================
# 14. DETAYLI DEĞERLENDİRME FONKSİYONU
# ============================================================

def evaluate_model(model, loader):
    model.eval()

    all_preds = []
    all_labels = []
    all_probs = []

    with torch.no_grad():
        for images, labels in loader:
            images = images.to(device)
            labels = labels.to(device)

            outputs = model(images)
            outputs = get_main_output(outputs)

            probs = torch.softmax(outputs, dim=1)[:, 1]
            _, preds = torch.max(outputs, 1)

            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())
            all_probs.extend(probs.cpu().numpy())

    all_labels = np.array(all_labels)
    all_preds = np.array(all_preds)
    all_probs = np.array(all_probs)

    metrics = {
        "accuracy": accuracy_score(all_labels, all_preds),
        "precision": precision_score(all_labels, all_preds, zero_division=0),
        "recall": recall_score(all_labels, all_preds, zero_division=0),
        "f1_score": f1_score(all_labels, all_preds, zero_division=0)
    }

    try:
        metrics["roc_auc"] = roc_auc_score(all_labels, all_probs)
    except ValueError:
        metrics["roc_auc"] = np.nan

    return metrics, all_labels, all_preds, all_probs


# ============================================================
# 15. GRAFİK FONKSİYONLARI
# ============================================================

def save_accuracy_graph(histories, model_display_name, save_path):
    avg_train_acc = np.mean(
        [h["train_acc"] for h in histories],
        axis=0
    )

    avg_val_acc = np.mean(
        [h["val_acc"] for h in histories],
        axis=0
    )

    epochs_range = range(1, len(avg_train_acc) + 1)

    plt.figure(figsize=(9, 6))
    plt.plot(
        epochs_range,
        avg_train_acc,
        marker="o",
        label="Average Train Accuracy"
    )
    plt.plot(
        epochs_range,
        avg_val_acc,
        marker="o",
        label="Average Validation Accuracy"
    )

    plt.title(f"{model_display_name} - Average Train vs Validation Accuracy")
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.savefig(save_path, dpi=300, bbox_inches="tight")
    plt.show()
    plt.close()


def save_loss_graph(histories, model_display_name, save_path):
    avg_train_loss = np.mean(
        [h["train_loss"] for h in histories],
        axis=0
    )

    avg_val_loss = np.mean(
        [h["val_loss"] for h in histories],
        axis=0
    )

    epochs_range = range(1, len(avg_train_loss) + 1)

    plt.figure(figsize=(9, 6))
    plt.plot(
        epochs_range,
        avg_train_loss,
        marker="o",
        label="Average Train Loss"
    )
    plt.plot(
        epochs_range,
        avg_val_loss,
        marker="o",
        label="Average Validation Loss"
    )

    plt.title(f"{model_display_name} - Average Train vs Validation Loss")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.savefig(save_path, dpi=300, bbox_inches="tight")
    plt.show()
    plt.close()


def save_confusion_matrix_graph(labels, preds, class_names, title, save_path):
    cm = confusion_matrix(labels, preds, labels=[0, 1])

    plt.figure(figsize=(6, 5))
    sns.heatmap(
        cm,
        annot=True,
        fmt="d",
        cmap="Blues",
        xticklabels=class_names,
        yticklabels=class_names
    )

    plt.title(title)
    plt.xlabel("Tahmin")
    plt.ylabel("Gerçek")
    plt.tight_layout()
    plt.savefig(save_path, dpi=300, bbox_inches="tight")
    plt.show()
    plt.close()


def save_roc_curve_graph(labels, probs, title, save_path):
    fpr, tpr, _ = roc_curve(labels, probs)
    auc_value = roc_auc_score(labels, probs)

    plt.figure(figsize=(7, 5))
    plt.plot(
        fpr,
        tpr,
        lw=2,
        label=f"ROC Curve - AUC = {auc_value:.2f}"
    )

    plt.plot(
        [0, 1],
        [0, 1],
        lw=2,
        linestyle="--",
        label="Random Classifier"
    )

    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title(title)
    plt.legend(loc="lower right")
    plt.grid(True)
    plt.tight_layout()
    plt.savefig(save_path, dpi=300, bbox_inches="tight")
    plt.show()
    plt.close()


# ============================================================
# 16. TEK MODEL BASELINE DENEYİ
# ============================================================

def run_cnn_baseline_experiment(model_name):
    config = MODEL_CONFIGS[model_name]

    model_display_name = config["display_name"]
    img_size = config["img_size"]

    print("\n" + "=" * 70)
    print(f"{model_display_name} deney başlıyor")
    print("=" * 70)

    model_output_root = os.path.join(
        OUTPUT_ROOT,
        model_name
    )

    charts_dir = os.path.join(
        model_output_root,
        "charts"
    )

    tables_dir = os.path.join(
        model_output_root,
        "tables"
    )

    os.makedirs(charts_dir, exist_ok=True)
    os.makedirs(tables_dir, exist_ok=True)

    baseline_transform = get_baseline_transforms(img_size)

    full_train_dataset = datasets.ImageFolder(
        os.path.join(DATA_DIR, "train"),
        transform=baseline_transform
    )

    test_dataset = datasets.ImageFolder(
        os.path.join(DATA_DIR, "test"),
        transform=baseline_transform
    )

    labels = np.array(full_train_dataset.targets)

    class_names = full_train_dataset.classes

    if len(class_names) == 2:
        display_class_names = ["Benign", "Malignant"]
    else:
        display_class_names = class_names

    print("Sınıflar:", full_train_dataset.class_to_idx)

    skf = StratifiedKFold(
        n_splits=N_SPLITS,
        shuffle=True,
        random_state=RANDOM_STATE
    )

    fold_results = []
    fold_summary_rows = []
    epoch_history_rows = []
    output_file_rows = []

    best_overall_val_acc = -1
    best_overall_state = None
    best_overall_fold = None
    best_overall_epoch = None

    criterion = nn.CrossEntropyLoss()

    for fold_no, (train_idx, val_idx) in enumerate(
        skf.split(np.zeros(len(labels)), labels),
        start=1
    ):
        print(f"\n--- {model_display_name} | FOLD {fold_no}/{N_SPLITS} ---")

        train_sub = Subset(
            full_train_dataset,
            train_idx
        )

        val_sub = Subset(
            full_train_dataset,
            val_idx
        )

        train_loader = DataLoader(
            train_sub,
            batch_size=BATCH_SIZE,
            shuffle=True,
            num_workers=2,
            pin_memory=torch.cuda.is_available()
        )

        val_loader = DataLoader(
            val_sub,
            batch_size=BATCH_SIZE,
            shuffle=False,
            num_workers=2,
            pin_memory=torch.cuda.is_available()
        )

        model = build_baseline_model(model_name)
        optimizer = get_baseline_optimizer(model)

        history = {
            "train_loss": [],
            "train_acc": [],
            "val_loss": [],
            "val_acc": []
        }

        best_epoch_val_acc = -1
        best_epoch_state = None
        best_epoch_no = 1

        for epoch in range(NUM_EPOCHS):
            train_loss, train_acc = train_one_epoch(
                model,
                train_loader,
                optimizer,
                criterion
            )

            val_loss, val_acc = evaluate_one_epoch(
                model,
                val_loader,
                criterion
            )

            history["train_loss"].append(train_loss)
            history["train_acc"].append(train_acc)
            history["val_loss"].append(val_loss)
            history["val_acc"].append(val_acc)

            epoch_history_rows.append({
                "Model": model_display_name,
                "Fold": fold_no,
                "Epoch": epoch + 1,
                "Train_Loss": train_loss,
                "Train_Accuracy": train_acc,
                "Validation_Loss": val_loss,
                "Validation_Accuracy": val_acc
            })

            print(
                f"Epoch {epoch+1}/{NUM_EPOCHS}: "
                f"Train Loss: {train_loss:.4f} | "
                f"Train Acc: {train_acc:.4f} | "
                f"Val Loss: {val_loss:.4f} | "
                f"Val Acc: {val_acc:.4f}"
            )

            if val_acc > best_epoch_val_acc:
                best_epoch_val_acc = val_acc
                best_epoch_no = epoch + 1
                best_epoch_state = clone_state_dict_to_cpu(model)

        if best_epoch_state is not None:
            model.load_state_dict(best_epoch_state)

        fold_results.append(history)

        val_metrics, val_labels, val_preds, val_probs = evaluate_model(
            model,
            val_loader
        )

        print(f"\n{model_display_name} Fold {fold_no} Validation Performansı")
        print(f"Best Epoch: {best_epoch_no}")
        print(f"Accuracy:  {val_metrics['accuracy']:.4f}")
        print(f"Precision: {val_metrics['precision']:.4f}")
        print(f"Recall:    {val_metrics['recall']:.4f}")
        print(f"F1-Score:  {val_metrics['f1_score']:.4f}")
        print(f"ROC-AUC:   {val_metrics['roc_auc']:.4f}")

        fold_summary_rows.append({
            "Model": model_display_name,
            "Fold": fold_no,
            "Best_Epoch": best_epoch_no,
            "Validation_Accuracy": val_metrics["accuracy"],
            "Validation_Precision": val_metrics["precision"],
            "Validation_Recall": val_metrics["recall"],
            "Validation_F1_Score": val_metrics["f1_score"],
            "Validation_ROC_AUC": val_metrics["roc_auc"]
        })

        if val_metrics["accuracy"] > best_overall_val_acc:
            best_overall_val_acc = val_metrics["accuracy"]
            best_overall_fold = fold_no
            best_overall_epoch = best_epoch_no
            best_overall_state = clone_state_dict_to_cpu(model)

        del model
        gc.collect()

        if torch.cuda.is_available():
            torch.cuda.empty_cache()

    print(f"\n{model_display_name} tüm foldlar tamamlandı.")

    # ========================================================
    # FINAL TEST
    # ========================================================

    test_loader = DataLoader(
        test_dataset,
        batch_size=BATCH_SIZE,
        shuffle=False,
        num_workers=2,
        pin_memory=torch.cuda.is_available()
    )

    final_model = build_baseline_model(model_name)
    final_model.load_state_dict(best_overall_state)
    final_model.to(device)

    test_metrics, test_labels, test_preds, test_probs = evaluate_model(
        final_model,
        test_loader
    )

    final_model_name = (
        f"{model_display_name} | Best Fold {best_overall_fold} | "
        f"Best Epoch {best_overall_epoch}"
    )

    print("\n--- FINAL TEST PERFORMANSI ---")
    print(f"Model:     {final_model_name}")
    print(f"Accuracy:  {test_metrics['accuracy']:.4f}")
    print(f"Precision: {test_metrics['precision']:.4f}")
    print(f"Recall:    {test_metrics['recall']:.4f}")
    print(f"F1-Score:  {test_metrics['f1_score']:.4f}")
    print(f"ROC-AUC:   {test_metrics['roc_auc']:.4f}")

    final_test_df = pd.DataFrame([{
        "Model": model_display_name,
        "Protocol": "CNN Baseline - pretrained backbone frozen, classifier only",
        "Final_Model": final_model_name,
        "Best_Fold": best_overall_fold,
        "Best_Epoch": best_overall_epoch,
        "Test_Accuracy": test_metrics["accuracy"],
        "Test_Precision": test_metrics["precision"],
        "Test_Recall": test_metrics["recall"],
        "Test_F1_Score": test_metrics["f1_score"],
        "Test_ROC_AUC": test_metrics["roc_auc"]
    }])

    # ========================================================
    # GRAFİKLER
    # ========================================================

    accuracy_graph_path = os.path.join(
        charts_dir,
        f"{model_name}_average_accuracy.png"
    )

    loss_graph_path = os.path.join(
        charts_dir,
        f"{model_name}_average_loss.png"
    )

    cm_graph_path = os.path.join(
        charts_dir,
        f"{model_name}_final_test_confusion_matrix.png"
    )

    roc_graph_path = os.path.join(
        charts_dir,
        f"{model_name}_final_test_roc_curve.png"
    )

    save_accuracy_graph(
        fold_results,
        model_display_name,
        accuracy_graph_path
    )

    save_loss_graph(
        fold_results,
        model_display_name,
        loss_graph_path
    )

    save_confusion_matrix_graph(
        test_labels,
        test_preds,
        display_class_names,
        title=f"{model_display_name} - Final Test Confusion Matrix",
        save_path=cm_graph_path
    )

    save_roc_curve_graph(
        test_labels,
        test_probs,
        title=f"{model_display_name} - Final Test ROC Curve",
        save_path=roc_graph_path
    )

    output_file_rows.extend([
        {
            "Model": model_display_name,
            "Output_Type": "Average Accuracy Graph",
            "File_Path": accuracy_graph_path
        },
        {
            "Model": model_display_name,
            "Output_Type": "Average Loss Graph",
            "File_Path": loss_graph_path
        },
        {
            "Model": model_display_name,
            "Output_Type": "Final Test Confusion Matrix Graph",
            "File_Path": cm_graph_path
        },
        {
            "Model": model_display_name,
            "Output_Type": "Final Test ROC Curve Graph",
            "File_Path": roc_graph_path
        }
    ])

    # ========================================================
    # EXCEL KAYIT
    # ========================================================

    fold_summary_df = pd.DataFrame(fold_summary_rows)
    epoch_history_df = pd.DataFrame(epoch_history_rows)
    output_files_df = pd.DataFrame(output_file_rows)

    excel_path = os.path.join(
        tables_dir,
        config["excel_name"]
    )

    with pd.ExcelWriter(excel_path, engine="openpyxl") as writer:
        fold_summary_df.to_excel(
            writer,
            sheet_name="Fold_Summary",
            index=False
        )

        epoch_history_df.to_excel(
            writer,
            sheet_name="Epoch_History",
            index=False
        )

        final_test_df.to_excel(
            writer,
            sheet_name="Final_Test",
            index=False
        )

        output_files_df.to_excel(
            writer,
            sheet_name="Output_Files",
            index=False
        )

    print("\n--- KAYIT TAMAMLANDI ---")
    print(f"Grafikler klasörü: {charts_dir}")
    print(f"Excel dosyası: {excel_path}")

    display(final_test_df)

    # Combined summary için geri dönüş
    summary_row = {
        "Model": model_display_name,
        "Protocol": "Baseline",
        "Best_Fold": best_overall_fold,
        "Best_Epoch": best_overall_epoch,
        "Test_Accuracy": test_metrics["accuracy"],
        "Test_Precision": test_metrics["precision"],
        "Test_Recall": test_metrics["recall"],
        "Test_F1_Score": test_metrics["f1_score"],
        "Test_ROC_AUC": test_metrics["roc_auc"],
        "Excel_Path": excel_path
    }

    del final_model
    gc.collect()

    if torch.cuda.is_available():
        torch.cuda.empty_cache()

    return summary_row


# ============================================================
# 17. TÜM CNN BASELINE MODELLERİNİ ÇALIŞTIRMA
# ============================================================

for model_name in RUN_MODELS:
    summary = run_cnn_baseline_experiment(model_name)
    combined_summary_rows.append(summary)


# ============================================================
# 18. TÜM BASELINE SONUÇLARINI TEK EXCEL'E KAYDETME
# ============================================================

combined_summary_df = pd.DataFrame(combined_summary_rows)

combined_excel_path = os.path.join(
    OUTPUT_ROOT,
    "cnn_baseline_combined_summary.xlsx"
)

with pd.ExcelWriter(combined_excel_path, engine="openpyxl") as writer:
    combined_summary_df.to_excel(
        writer,
        sheet_name="CNN_Baseline_Summary",
        index=False
    )

print("\n" + "=" * 70)
print("TÜM CNN BASELINE DENEYLERİ TAMAMLANDI")
print("=" * 70)

print(f"Toplu özet Excel dosyası: {combined_excel_path}")

display(combined_summary_df)
```
