```python
# ============================================================
# ViT-L/16 Optimized
# 5-Fold Cross Validation + Her Fold Train / Validation / Test Sonucu
# ============================================================

import os
import copy
import random
import numpy as np
import pandas as pd

import torch
import torch.nn as nn
import torch.optim as optim

from torch.utils.data import DataLoader, Subset
from torchvision import datasets, transforms, models

from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
    confusion_matrix,
    roc_curve
)
from sklearn.utils.class_weight import compute_class_weight

import matplotlib.pyplot as plt

from google.colab import drive


# ============================================================
# 1. DENEY AYARLARI
# ============================================================

MODEL_NAME = "ViT-L/16"
EXPERIMENT_NAME = "Optimized"
OPTIMIZED = True

DATA_DIR = "/content/breast_mri_dataset"
BATCH_SIZE = 32
NUM_EPOCHS = 10
IMG_SIZE = 224
NUM_CLASSES = 2
K_FOLDS = 5

PATIENCE = 5
GRAD_CLIP = 1.0

OUTPUT_DIR = "/content/vit_l16_optimized_outputs"
os.makedirs(OUTPUT_DIR, exist_ok=True)

BLUE = "#1f77b4"
LIGHT_BLUE = "#6baed6"
DARK_BLUE = "#084594"
WHITE = "white"
BLACK = "black"


def set_seed(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    torch.backends.cudnn.deterministic = False
    torch.backends.cudnn.benchmark = True


set_seed(42)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Cihaz: {device}")
print(f"Deney: {EXPERIMENT_NAME}")
print(f"Model: {MODEL_NAME}")
print(f"5 Fold: {K_FOLDS}, Epoch: {NUM_EPOCHS}, Batch Size: {BATCH_SIZE}")

plt.rcParams["figure.facecolor"] = "white"
plt.rcParams["axes.facecolor"] = "white"
plt.rcParams["savefig.facecolor"] = "white"
plt.rcParams["axes.edgecolor"] = "black"
plt.rcParams["axes.labelcolor"] = "black"
plt.rcParams["xtick.color"] = "black"
plt.rcParams["ytick.color"] = "black"
plt.rcParams["text.color"] = "black"
plt.rcParams["legend.frameon"] = True


# ============================================================
# 2. DRIVE BAĞLAMA VE DATASET AÇMA
# ============================================================

drive.mount('/content/drive')

!cp "/content/drive/MyDrive/data_set_zip/breast_mri_dataset.rar" /content/
!unrar x -o+ /content/breast_mri_dataset.rar /content/


# ============================================================
# 3. TRANSFORMASYONLAR
# ============================================================
# Equalized baseline:
# - Train augmentation var.
# - Validation/Test augmentation yok.
# - Adam + CrossEntropyLoss.
# - Scheduler, early stopping, class weight, label smoothing, mixed precision yok.
#
# Optimized:
# - Train augmentation var.
# - RandomErasing var.
# - Son 4 encoder layer fine-tuning.
# - AdamW + scheduler + class weight + label smoothing + mixed precision + early stopping.

vit_mean = [0.485, 0.456, 0.406]
vit_std = [0.229, 0.224, 0.225]

train_transform_items = [
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.RandomRotation(degrees=15),
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.RandomVerticalFlip(p=0.2),
    transforms.RandomAffine(
        degrees=0,
        translate=(0.05, 0.05),
        scale=(0.95, 1.05)
    ),
    transforms.ColorJitter(
        brightness=0.15,
        contrast=0.15
    ),
    transforms.ToTensor(),
    transforms.Normalize(mean=vit_mean, std=vit_std)
]

if OPTIMIZED:
    train_transform_items.append(
        transforms.RandomErasing(
            p=0.15,
            scale=(0.02, 0.08),
            ratio=(0.3, 3.3)
        )
    )

train_transform = transforms.Compose(train_transform_items)

eval_transform = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.ToTensor(),
    transforms.Normalize(mean=vit_mean, std=vit_std)
])


# ============================================================
# 4. DATASETLER
# ============================================================

train_aug_dataset = datasets.ImageFolder(
    os.path.join(DATA_DIR, "train"),
    transform=train_transform
)

train_eval_dataset = datasets.ImageFolder(
    os.path.join(DATA_DIR, "train"),
    transform=eval_transform
)

test_dataset = datasets.ImageFolder(
    os.path.join(DATA_DIR, "test"),
    transform=eval_transform
)

class_names = train_eval_dataset.classes
print("Sınıflar:", train_eval_dataset.class_to_idx)

if len(class_names) == 2:
    display_class_names = ["Benign", "Malignant"]
else:
    display_class_names = class_names


test_loader = DataLoader(
    test_dataset,
    batch_size=BATCH_SIZE,
    shuffle=False,
    num_workers=2,
    pin_memory=torch.cuda.is_available()
)


# ============================================================
# 5. ViT-L/16 MODEL
# ============================================================

def build_model():
    model = models.vit_l_16(
        weights=models.ViT_L_16_Weights.IMAGENET1K_V1
    )

    for param in model.parameters():
        param.requires_grad = False

    if OPTIMIZED:
        trainable_keywords = [
            "encoder.layers.encoder_layer_20",
            "encoder.layers.encoder_layer_21",
            "encoder.layers.encoder_layer_22",
            "encoder.layers.encoder_layer_23",
            "encoder.ln"
        ]
    else:
        trainable_keywords = [
            "encoder.layers.encoder_layer_22",
            "encoder.layers.encoder_layer_23",
            "encoder.ln"
        ]

    for name, param in model.named_parameters():
        if any(key in name for key in trainable_keywords):
            param.requires_grad = True

    in_features = model.heads.head.in_features

    if OPTIMIZED:
        model.heads.head = nn.Sequential(
            nn.Dropout(p=0.3),
            nn.Linear(in_features, NUM_CLASSES)
        )
    else:
        model.heads.head = nn.Linear(in_features, NUM_CLASSES)

    return model.to(device)


# ============================================================
# 6. TRAIN / EVALUATE / PREDICT
# ============================================================

scaler = torch.cuda.amp.GradScaler(enabled=(OPTIMIZED and torch.cuda.is_available()))


def train_one_epoch(model, loader, optimizer, criterion):
    model.train()

    running_loss = 0.0
    correct = 0
    total = 0

    for images, labels in loader:
        images = images.to(device)
        labels = labels.to(device)

        optimizer.zero_grad()

        with torch.cuda.amp.autocast(enabled=(OPTIMIZED and torch.cuda.is_available())):
            outputs = model(images)
            loss = criterion(outputs, labels)

        if OPTIMIZED:
            scaler.scale(loss).backward()
            scaler.unscale_(optimizer)
            torch.nn.utils.clip_grad_norm_(
                filter(lambda p: p.requires_grad, model.parameters()),
                GRAD_CLIP
            )
            scaler.step(optimizer)
            scaler.update()
        else:
            loss.backward()
            optimizer.step()

        running_loss += loss.item() * images.size(0)

        _, preds = torch.max(outputs, 1)
        correct += torch.sum(preds == labels).item()
        total += labels.size(0)

    epoch_loss = running_loss / total
    epoch_acc = correct / total

    return epoch_loss, epoch_acc


def evaluate(model, loader, criterion):
    model.eval()

    running_loss = 0.0
    correct = 0
    total = 0

    with torch.no_grad():
        for images, labels in loader:
            images = images.to(device)
            labels = labels.to(device)

            with torch.cuda.amp.autocast(enabled=(OPTIMIZED and torch.cuda.is_available())):
                outputs = model(images)
                loss = criterion(outputs, labels)

            running_loss += loss.item() * images.size(0)

            _, preds = torch.max(outputs, 1)
            correct += torch.sum(preds == labels).item()
            total += labels.size(0)

    epoch_loss = running_loss / total
    epoch_acc = correct / total

    return epoch_loss, epoch_acc


def get_predictions(model, loader):
    model.eval()

    all_labels = []
    all_preds = []
    all_probs = []

    with torch.no_grad():
        for images, labels in loader:
            images = images.to(device)
            labels = labels.to(device)

            with torch.cuda.amp.autocast(enabled=(OPTIMIZED and torch.cuda.is_available())):
                outputs = model(images)

            probs = torch.softmax(outputs, dim=1)

            if NUM_CLASSES == 2:
                positive_probs = probs[:, 1]
            else:
                positive_probs = torch.max(probs, dim=1).values

            _, preds = torch.max(outputs, 1)

            all_labels.extend(labels.cpu().numpy())
            all_preds.extend(preds.cpu().numpy())
            all_probs.extend(positive_probs.cpu().numpy())

    return np.array(all_labels), np.array(all_preds), np.array(all_probs)


# ============================================================
# 7. METRİKLER VE GRAFİKLER
# ============================================================

def calculate_metrics(labels, preds, probs):
    acc = accuracy_score(labels, preds)
    precision = precision_score(labels, preds, zero_division=0)
    recall = recall_score(labels, preds, zero_division=0)
    f1 = f1_score(labels, preds, zero_division=0)

    try:
        auc = roc_auc_score(labels, probs)
    except ValueError:
        auc = 0.0

    return {
        "Accuracy": acc,
        "Precision": precision,
        "Recall": recall,
        "F1-Score": f1,
        "ROC-AUC": auc
    }


def save_and_show_plot(filename):
    path = os.path.join(OUTPUT_DIR, filename)
    plt.tight_layout()
    plt.savefig(path, dpi=300, bbox_inches="tight", facecolor="white")
    plt.show()
    plt.close()


def plot_accuracy(history, fold):
    epochs_range = range(1, len(history["train_acc"]) + 1)

    plt.figure(figsize=(9, 6), facecolor="white")
    ax = plt.gca()
    ax.set_facecolor("white")

    plt.plot(
        epochs_range,
        history["train_acc"],
        marker="o",
        color=BLUE,
        linewidth=2,
        label="Train Accuracy"
    )

    plt.plot(
        epochs_range,
        history["val_acc"],
        marker="o",
        color=LIGHT_BLUE,
        linewidth=2,
        label="Validation Accuracy"
    )

    plt.title(f"{MODEL_NAME} {EXPERIMENT_NAME} - Fold {fold} - Accuracy")
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.ylim(0, 1)
    plt.legend()
    plt.grid(True, alpha=0.3)

    save_and_show_plot(f"vit_l16_{EXPERIMENT_NAME.lower().replace(' ', '_')}_fold_{fold}_train_val_accuracy.png")


def plot_loss(history, fold):
    epochs_range = range(1, len(history["train_loss"]) + 1)

    plt.figure(figsize=(9, 6), facecolor="white")
    ax = plt.gca()
    ax.set_facecolor("white")

    plt.plot(
        epochs_range,
        history["train_loss"],
        marker="o",
        color=BLUE,
        linewidth=2,
        label="Train Loss"
    )

    plt.plot(
        epochs_range,
        history["val_loss"],
        marker="o",
        color=LIGHT_BLUE,
        linewidth=2,
        label="Validation Loss"
    )

    plt.title(f"{MODEL_NAME} {EXPERIMENT_NAME} - Fold {fold} - Loss")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.legend()
    plt.grid(True, alpha=0.3)

    save_and_show_plot(f"vit_l16_{EXPERIMENT_NAME.lower().replace(' ', '_')}_fold_{fold}_train_val_loss.png")


def plot_confusion_matrix(labels, preds, title, filename):
    cm = confusion_matrix(labels, preds)

    plt.figure(figsize=(6, 5), facecolor="white")
    ax = plt.gca()
    ax.set_facecolor("white")

    im = plt.imshow(cm, interpolation="nearest", cmap=plt.cm.Blues)

    plt.title(title)
    plt.colorbar(im)

    tick_marks = np.arange(len(display_class_names))
    plt.xticks(tick_marks, display_class_names, rotation=45)
    plt.yticks(tick_marks, display_class_names)

    thresh = cm.max() / 2.0 if cm.max() > 0 else 0

    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            plt.text(
                j,
                i,
                format(cm[i, j], "d"),
                ha="center",
                va="center",
                color=WHITE if cm[i, j] > thresh else BLACK,
                fontsize=12,
                fontweight="bold"
            )

    plt.ylabel("Gerçek")
    plt.xlabel("Tahmin")

    save_and_show_plot(filename)


def plot_test_metrics(metrics, fold):
    metric_names = list(metrics.keys())
    metric_values = list(metrics.values())

    plt.figure(figsize=(8, 5), facecolor="white")
    ax = plt.gca()
    ax.set_facecolor("white")

    bars = plt.bar(
        metric_names,
        metric_values,
        color=BLUE,
        edgecolor=DARK_BLUE
    )

    plt.ylim(0, 1)
    plt.title(f"{MODEL_NAME} {EXPERIMENT_NAME} - Fold {fold} - Test Metrics")
    plt.xlabel("Metric")
    plt.ylabel("Score")
    plt.grid(axis="y", alpha=0.3)

    for bar, value in zip(bars, metric_values):
        plt.text(
            bar.get_x() + bar.get_width() / 2,
            value + 0.02,
            f"{value:.3f}",
            ha="center",
            va="bottom",
            color=BLACK,
            fontweight="bold"
        )

    save_and_show_plot(f"vit_l16_{EXPERIMENT_NAME.lower().replace(' ', '_')}_fold_{fold}_test_metrics.png")


def plot_roc_curve(labels, probs, auc_score, fold, split_name):
    try:
        fpr, tpr, _ = roc_curve(labels, probs)
    except ValueError:
        print(f"Fold {fold}: {split_name} ROC Curve çizilemedi. Tek sınıf olabilir.")
        return

    plt.figure(figsize=(7, 5), facecolor="white")
    ax = plt.gca()
    ax.set_facecolor("white")

    plt.plot(
        fpr,
        tpr,
        color=BLUE,
        lw=2,
        label=f"ROC Curve - AUC = {auc_score:.3f}"
    )

    plt.plot(
        [0, 1],
        [0, 1],
        color=LIGHT_BLUE,
        lw=2,
        linestyle="--",
        label="Random Classifier"
    )

    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title(f"{MODEL_NAME} {EXPERIMENT_NAME} - Fold {fold} - {split_name} ROC Curve")
    plt.legend(loc="lower right")
    plt.grid(True, alpha=0.3)

    save_and_show_plot(f"vit_l16_{EXPERIMENT_NAME.lower().replace(' ', '_')}_fold_{fold}_{split_name.lower()}_roc_curve.png")


# ============================================================
# 8. CLASS WEIGHT
# ============================================================

targets = np.array([label for _, label in train_eval_dataset.samples])

if OPTIMIZED:
    classes = np.unique(targets)
    class_weights = compute_class_weight(
        class_weight="balanced",
        classes=classes,
        y=targets
    )
    class_weights = torch.tensor(class_weights, dtype=torch.float).to(device)
    print("Class weights:", class_weights)
else:
    class_weights = None


# ============================================================
# 9. 5-FOLD CROSS VALIDATION
# ============================================================

skf = StratifiedKFold(
    n_splits=K_FOLDS,
    shuffle=True,
    random_state=42
)

fold_train_results = []
fold_validation_results = []
fold_test_results = []

print(f"\n--- {MODEL_NAME} {EXPERIMENT_NAME} 5-Fold Cross Validation Başlıyor ---")

for fold, (train_idx, val_idx) in enumerate(skf.split(np.zeros(len(targets)), targets), 1):

    print("\n" + "=" * 80)
    print(f"{MODEL_NAME} {EXPERIMENT_NAME} FOLD {fold}/{K_FOLDS}")
    print("=" * 80)

    train_subset = Subset(train_aug_dataset, train_idx)
    train_eval_subset = Subset(train_eval_dataset, train_idx)
    val_subset = Subset(train_eval_dataset, val_idx)

    train_loader = DataLoader(
        train_subset,
        batch_size=BATCH_SIZE,
        shuffle=True,
        num_workers=2,
        pin_memory=torch.cuda.is_available()
    )

    train_eval_loader = DataLoader(
        train_eval_subset,
        batch_size=BATCH_SIZE,
        shuffle=False,
        num_workers=2,
        pin_memory=torch.cuda.is_available()
    )

    val_loader = DataLoader(
        val_subset,
        batch_size=BATCH_SIZE,
        shuffle=False,
        num_workers=2,
        pin_memory=torch.cuda.is_available()
    )

    model = build_model()

    if OPTIMIZED:
        criterion = nn.CrossEntropyLoss(
            weight=class_weights,
            label_smoothing=0.05
        )
        optimizer = optim.AdamW(
            filter(lambda p: p.requires_grad, model.parameters()),
            lr=3e-5,
            weight_decay=1e-4
        )
        scheduler = optim.lr_scheduler.CosineAnnealingLR(
            optimizer,
            T_max=NUM_EPOCHS,
            eta_min=1e-6
        )
    else:
        criterion = nn.CrossEntropyLoss()
        optimizer = optim.Adam(
            filter(lambda p: p.requires_grad, model.parameters()),
            lr=1e-4
        )
        scheduler = None

    history = {
        "train_loss": [],
        "train_acc": [],
        "val_loss": [],
        "val_acc": []
    }

    best_val_loss = float("inf")
    best_model_wts = copy.deepcopy(model.state_dict())
    patience_counter = 0
    best_epoch = NUM_EPOCHS

    for epoch in range(NUM_EPOCHS):
        train_loss, train_acc = train_one_epoch(
            model,
            train_loader,
            optimizer,
            criterion
        )

        val_loss, val_acc = evaluate(
            model,
            val_loader,
            criterion
        )

        if scheduler is not None:
            scheduler.step()

        history["train_loss"].append(train_loss)
        history["train_acc"].append(train_acc)
        history["val_loss"].append(val_loss)
        history["val_acc"].append(val_acc)

        lr_text = optimizer.param_groups[0]["lr"]

        print(
            f"Fold {fold} | Epoch {epoch+1}/{NUM_EPOCHS} | "
            f"Train Loss: {train_loss:.4f} | Train Acc: {train_acc:.4f} | "
            f"Val Loss: {val_loss:.4f} | Val Acc: {val_acc:.4f} | "
            f"LR: {lr_text:.8f}"
        )

        if OPTIMIZED:
            if val_loss < best_val_loss:
                best_val_loss = val_loss
                best_model_wts = copy.deepcopy(model.state_dict())
                best_epoch = epoch + 1
                patience_counter = 0
            else:
                patience_counter += 1

            if patience_counter >= PATIENCE:
                print(f"Early stopping aktif. Fold {fold}, Best Epoch: {best_epoch}")
                break

    if OPTIMIZED:
        model.load_state_dict(best_model_wts)

    plot_accuracy(history, fold)
    plot_loss(history, fold)

    # Train sonuçları
    train_labels, train_preds, train_probs = get_predictions(model, train_eval_loader)
    train_metrics = calculate_metrics(train_labels, train_preds, train_probs)

    print(f"\n--- FOLD {fold} TRAIN SETİ SONUÇLARI ---")
    for metric_name, metric_value in train_metrics.items():
        print(f"{metric_name}: {metric_value:.4f}")

    plot_confusion_matrix(
        train_labels,
        train_preds,
        f"{MODEL_NAME} {EXPERIMENT_NAME} - Fold {fold} - Train Confusion Matrix",
        f"vit_l16_{EXPERIMENT_NAME.lower().replace(' ', '_')}_fold_{fold}_train_confusion_matrix.png"
    )

    plot_roc_curve(
        train_labels,
        train_probs,
        train_metrics["ROC-AUC"],
        fold,
        "Train"
    )

    train_row = {
        "Fold": fold,
        "Final Train Loss": history["train_loss"][-1],
        "Final Train Accuracy": history["train_acc"][-1],
        **train_metrics
    }

    if OPTIMIZED:
        train_row["Best Epoch"] = best_epoch

    fold_train_results.append(train_row)

    # Validation sonuçları
    val_labels, val_preds, val_probs = get_predictions(model, val_loader)
    val_metrics = calculate_metrics(val_labels, val_preds, val_probs)

    print(f"\n--- FOLD {fold} VALIDATION SETİ SONUÇLARI ---")
    for metric_name, metric_value in val_metrics.items():
        print(f"{metric_name}: {metric_value:.4f}")

    plot_confusion_matrix(
        val_labels,
        val_preds,
        f"{MODEL_NAME} {EXPERIMENT_NAME} - Fold {fold} - Validation Confusion Matrix",
        f"vit_l16_{EXPERIMENT_NAME.lower().replace(' ', '_')}_fold_{fold}_validation_confusion_matrix.png"
    )

    plot_roc_curve(
        val_labels,
        val_probs,
        val_metrics["ROC-AUC"],
        fold,
        "Validation"
    )

    validation_row = {
        "Fold": fold,
        "Final Validation Loss": history["val_loss"][-1],
        "Final Validation Accuracy": history["val_acc"][-1],
        **val_metrics
    }

    if OPTIMIZED:
        validation_row["Best Epoch"] = best_epoch
        validation_row["Best Validation Loss"] = best_val_loss

    fold_validation_results.append(validation_row)

    # Test sonuçları
    test_labels, test_preds, test_probs = get_predictions(model, test_loader)
    test_metrics = calculate_metrics(test_labels, test_preds, test_probs)

    print(f"\n--- FOLD {fold} TEST SETİ SONUÇLARI ---")
    for metric_name, metric_value in test_metrics.items():
        print(f"{metric_name}: {metric_value:.4f}")

    test_row = {
        "Fold": fold,
        **test_metrics
    }

    if OPTIMIZED:
        test_row["Best Epoch"] = best_epoch

    fold_test_results.append(test_row)

    plot_test_metrics(test_metrics, fold)

    plot_confusion_matrix(
        test_labels,
        test_preds,
        f"{MODEL_NAME} {EXPERIMENT_NAME} - Fold {fold} - Test Confusion Matrix",
        f"vit_l16_{EXPERIMENT_NAME.lower().replace(' ', '_')}_fold_{fold}_test_confusion_matrix.png"
    )

    plot_roc_curve(
        test_labels,
        test_probs,
        test_metrics["ROC-AUC"],
        fold,
        "Test"
    )

    torch.save(
        model.state_dict(),
        os.path.join(
            OUTPUT_DIR,
            f"vit_l16_{EXPERIMENT_NAME.lower().replace(' ', '_')}_fold_{fold}.pth"
        )
    )


# ============================================================
# 10. TÜM FOLD SONUÇLARI
# ============================================================

train_results_df = pd.DataFrame(fold_train_results)
validation_results_df = pd.DataFrame(fold_validation_results)
test_results_df = pd.DataFrame(fold_test_results)

print("\n" + "=" * 80)
print(f"{MODEL_NAME} {EXPERIMENT_NAME} TÜM FOLD TRAIN SONUÇLARI")
print("=" * 80)
print(train_results_df)

print("\n" + "=" * 80)
print(f"{MODEL_NAME} {EXPERIMENT_NAME} TÜM FOLD VALIDATION SONUÇLARI")
print("=" * 80)
print(validation_results_df)

print("\n" + "=" * 80)
print(f"{MODEL_NAME} {EXPERIMENT_NAME} TÜM FOLD TEST SONUÇLARI")
print("=" * 80)
print(test_results_df)

print("\n" + "=" * 80)
print(f"{MODEL_NAME} {EXPERIMENT_NAME} TEST ORTALAMA ± STD")
print("=" * 80)

for col in ["Accuracy", "Precision", "Recall", "F1-Score", "ROC-AUC"]:
    mean_val = test_results_df[col].mean()
    std_val = test_results_df[col].std()
    print(f"{col}: {mean_val:.4f} ± {std_val:.4f}")

file_prefix = f"vit_l16_{EXPERIMENT_NAME.lower().replace(' ', '_')}"

train_results_df.to_csv(
    os.path.join(OUTPUT_DIR, f"{file_prefix}_train_results.csv"),
    index=False
)

validation_results_df.to_csv(
    os.path.join(OUTPUT_DIR, f"{file_prefix}_validation_results.csv"),
    index=False
)

test_results_df.to_csv(
    os.path.join(OUTPUT_DIR, f"{file_prefix}_test_results.csv"),
    index=False
)

print(f"\nÇıktılar kaydedildi: {OUTPUT_DIR}")
```
