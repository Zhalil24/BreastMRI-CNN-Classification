```python
# ============================================================
# ViT-L/32 BASELINE
# 5-Fold Cross Validation + Her Fold Train / Validation / Test Sonucu
#
# Koşullar:
# - 5 Fold
# - 10 Epoch
# - Batch Size 32
# - ImageNet pretrained
# - Train data augmentation VAR
# - Validation/Test augmentation YOK
# - Son encoder blokları + head trainable
# - Adam optimizer
# - CrossEntropyLoss
# - Scheduler YOK
# - Early stopping YOK
# - Label smoothing YOK
# - Class weight YOK
# - Mixed precision YOK
# ============================================================


# ============================================================
# 1. KÜTÜPHANELER
# ============================================================

import os
import copy
import random
import numpy as np
import pandas as pd

import torch
import torch.nn as nn
import torch.optim as optim

from torch.utils.data import DataLoader, Subset
from torchvision import datasets, transforms, models

from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
    confusion_matrix,
    roc_curve
)

import matplotlib.pyplot as plt

from google.colab import drive


# ============================================================
# 2. SEED
# ============================================================

def set_seed(seed=42):
    random.seed(seed)
    np.random.seed(seed)

    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)

    torch.backends.cudnn.deterministic = False
    torch.backends.cudnn.benchmark = True


set_seed(42)


# ============================================================
# 3. GRAFİK GENEL TEMA
# ============================================================

plt.rcParams["figure.facecolor"] = "white"
plt.rcParams["axes.facecolor"] = "white"
plt.rcParams["savefig.facecolor"] = "white"
plt.rcParams["axes.edgecolor"] = "black"
plt.rcParams["axes.labelcolor"] = "black"
plt.rcParams["xtick.color"] = "black"
plt.rcParams["ytick.color"] = "black"
plt.rcParams["text.color"] = "black"
plt.rcParams["legend.frameon"] = True


# ============================================================
# 4. DRIVE BAĞLAMA VE DATASET AÇMA
# ============================================================

drive.mount('/content/drive')

!cp "/content/drive/MyDrive/data_set_zip/breast_mri_dataset.rar" /content/
!unrar x -o+ /content/breast_mri_dataset.rar /content/


# ============================================================
# 5. AYARLAR
# ============================================================

DATA_DIR = "/content/breast_mri_dataset"

BATCH_SIZE = 32
NUM_EPOCHS = 10
IMG_SIZE = 224
NUM_CLASSES = 2
K_FOLDS = 5

OUTPUT_DIR = "/content/vit_l32_baseline_outputs"
os.makedirs(OUTPUT_DIR, exist_ok=True)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Cihaz: {device}")


# ============================================================
# 6. RENK AYARLARI
# ============================================================

BLUE = "#1f77b4"
LIGHT_BLUE = "#6baed6"
DARK_BLUE = "#084594"
WHITE = "white"
BLACK = "black"


# ============================================================
# 7. TRANSFORMASYONLAR
# ============================================================

vit_mean = [0.485, 0.456, 0.406]
vit_std = [0.229, 0.224, 0.225]

train_transform = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),

    transforms.RandomRotation(degrees=15),
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.RandomVerticalFlip(p=0.2),

    transforms.RandomAffine(
        degrees=0,
        translate=(0.05, 0.05),
        scale=(0.95, 1.05)
    ),

    transforms.ColorJitter(
        brightness=0.15,
        contrast=0.15
    ),

    transforms.ToTensor(),
    transforms.Normalize(mean=vit_mean, std=vit_std)
])

eval_transform = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.ToTensor(),
    transforms.Normalize(mean=vit_mean, std=vit_std)
])


# ============================================================
# 8. DATASETLER
# ============================================================

full_train_aug_dataset = datasets.ImageFolder(
    os.path.join(DATA_DIR, "train"),
    transform=train_transform
)

full_train_eval_dataset = datasets.ImageFolder(
    os.path.join(DATA_DIR, "train"),
    transform=eval_transform
)

test_dataset = datasets.ImageFolder(
    os.path.join(DATA_DIR, "test"),
    transform=eval_transform
)

class_names = full_train_eval_dataset.classes
print("Sınıflar:", full_train_eval_dataset.class_to_idx)

if len(class_names) == 2:
    display_class_names = ["Benign", "Malignant"]
else:
    display_class_names = class_names


# ============================================================
# 9. TEST LOADER
# ============================================================

test_loader = DataLoader(
    test_dataset,
    batch_size=BATCH_SIZE,
    shuffle=False,
    num_workers=2,
    pin_memory=torch.cuda.is_available()
)


# ============================================================
# 10. ViT-L/32 MODEL
# ============================================================

def build_vit_l32_baseline_model():
    model = models.vit_l_32(
        weights=models.ViT_L_32_Weights.IMAGENET1K_V1
    )

    # Önce tüm parametreleri dondur.
    for param in model.parameters():
        param.requires_grad = False

    # ViT-L/32 24 encoder layer içerir.
    # Baseline eşitleme için son 2 encoder layer ve encoder.ln açılır.
    trainable_keywords = [
        "encoder.layers.encoder_layer_22",
        "encoder.layers.encoder_layer_23",
        "encoder.ln"
    ]

    for name, param in model.named_parameters():
        if any(key in name for key in trainable_keywords):
            param.requires_grad = True

    in_features = model.heads.head.in_features
    model.heads.head = nn.Linear(in_features, NUM_CLASSES)

    return model.to(device)


# ============================================================
# 11. TRAIN / EVALUATE / PREDICT
# ============================================================

def train_one_epoch(model, loader, optimizer, criterion):
    model.train()

    running_loss = 0.0
    correct = 0
    total = 0

    for images, labels in loader:
        images = images.to(device)
        labels = labels.to(device)

        optimizer.zero_grad()

        outputs = model(images)
        loss = criterion(outputs, labels)

        loss.backward()
        optimizer.step()

        running_loss += loss.item() * images.size(0)

        _, preds = torch.max(outputs, 1)
        correct += torch.sum(preds == labels).item()
        total += labels.size(0)

    return running_loss / total, correct / total


def evaluate(model, loader, criterion):
    model.eval()

    running_loss = 0.0
    correct = 0
    total = 0

    with torch.no_grad():
        for images, labels in loader:
            images = images.to(device)
            labels = labels.to(device)

            outputs = model(images)
            loss = criterion(outputs, labels)

            running_loss += loss.item() * images.size(0)

            _, preds = torch.max(outputs, 1)
            correct += torch.sum(preds == labels).item()
            total += labels.size(0)

    return running_loss / total, correct / total


def get_predictions(model, loader):
    model.eval()

    all_labels = []
    all_preds = []
    all_probs = []

    with torch.no_grad():
        for images, labels in loader:
            images = images.to(device)
            labels = labels.to(device)

            outputs = model(images)
            probs = torch.softmax(outputs, dim=1)

            if NUM_CLASSES == 2:
                positive_probs = probs[:, 1]
            else:
                positive_probs = torch.max(probs, dim=1).values

            _, preds = torch.max(outputs, 1)

            all_labels.extend(labels.cpu().numpy())
            all_preds.extend(preds.cpu().numpy())
            all_probs.extend(positive_probs.cpu().numpy())

    return np.array(all_labels), np.array(all_preds), np.array(all_probs)


# ============================================================
# 12. METRİK HESAPLAMA
# ============================================================

def calculate_metrics(labels, preds, probs):
    acc = accuracy_score(labels, preds)
    precision = precision_score(labels, preds, zero_division=0)
    recall = recall_score(labels, preds, zero_division=0)
    f1 = f1_score(labels, preds, zero_division=0)

    try:
        auc = roc_auc_score(labels, probs)
    except ValueError:
        auc = 0.0

    return {
        "Accuracy": acc,
        "Precision": precision,
        "Recall": recall,
        "F1-Score": f1,
        "ROC-AUC": auc
    }


# ============================================================
# 13. GRAFİK FONKSİYONLARI
# ============================================================

def save_and_show_plot(filename):
    path = os.path.join(OUTPUT_DIR, filename)
    plt.tight_layout()
    plt.savefig(path, dpi=300, bbox_inches="tight", facecolor="white")
    plt.show()
    plt.close()


def plot_accuracy(history, fold):
    epochs_range = range(1, len(history["train_acc"]) + 1)

    plt.figure(figsize=(9, 6), facecolor="white")
    ax = plt.gca()
    ax.set_facecolor("white")

    plt.plot(epochs_range, history["train_acc"], marker="o", color=BLUE, linewidth=2, label="Train Accuracy")
    plt.plot(epochs_range, history["val_acc"], marker="o", color=LIGHT_BLUE, linewidth=2, label="Validation Accuracy")

    plt.title(f"ViT-L/32 Baseline - Fold {fold} - Accuracy")
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.ylim(0, 1)
    plt.legend()
    plt.grid(True, alpha=0.3)

    save_and_show_plot(f"vit_l32_baseline_fold_{fold}_train_val_accuracy.png")


def plot_loss(history, fold):
    epochs_range = range(1, len(history["train_loss"]) + 1)

    plt.figure(figsize=(9, 6), facecolor="white")
    ax = plt.gca()
    ax.set_facecolor("white")

    plt.plot(epochs_range, history["train_loss"], marker="o", color=BLUE, linewidth=2, label="Train Loss")
    plt.plot(epochs_range, history["val_loss"], marker="o", color=LIGHT_BLUE, linewidth=2, label="Validation Loss")

    plt.title(f"ViT-L/32 Baseline - Fold {fold} - Loss")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.legend()
    plt.grid(True, alpha=0.3)

    save_and_show_plot(f"vit_l32_baseline_fold_{fold}_train_val_loss.png")


def plot_confusion_matrix(labels, preds, title, filename):
    cm = confusion_matrix(labels, preds)

    plt.figure(figsize=(6, 5), facecolor="white")
    ax = plt.gca()
    ax.set_facecolor("white")

    im = plt.imshow(cm, interpolation="nearest", cmap=plt.cm.Blues)

    plt.title(title)
    plt.colorbar(im)

    tick_marks = np.arange(len(display_class_names))
    plt.xticks(tick_marks, display_class_names, rotation=45)
    plt.yticks(tick_marks, display_class_names)

    thresh = cm.max() / 2.0 if cm.max() > 0 else 0

    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            plt.text(
                j,
                i,
                format(cm[i, j], "d"),
                ha="center",
                va="center",
                color=WHITE if cm[i, j] > thresh else BLACK,
                fontsize=12,
                fontweight="bold"
            )

    plt.ylabel("Gerçek")
    plt.xlabel("Tahmin")

    save_and_show_plot(filename)


def plot_test_metrics(metrics, fold):
    metric_names = list(metrics.keys())
    metric_values = list(metrics.values())

    plt.figure(figsize=(8, 5), facecolor="white")
    ax = plt.gca()
    ax.set_facecolor("white")

    bars = plt.bar(metric_names, metric_values, color=BLUE, edgecolor=DARK_BLUE)

    plt.ylim(0, 1)
    plt.title(f"ViT-L/32 Baseline - Fold {fold} - Test Metrics")
    plt.xlabel("Metric")
    plt.ylabel("Score")
    plt.grid(axis="y", alpha=0.3)

    for bar, value in zip(bars, metric_values):
        plt.text(
            bar.get_x() + bar.get_width() / 2,
            value + 0.02,
            f"{value:.3f}",
            ha="center",
            va="bottom",
            color=BLACK,
            fontweight="bold"
        )

    save_and_show_plot(f"vit_l32_baseline_fold_{fold}_test_metrics.png")


def plot_roc_curve(labels, probs, auc_score, fold, split_name):
    try:
        fpr, tpr, _ = roc_curve(labels, probs)
    except ValueError:
        print(f"Fold {fold}: {split_name} ROC Curve çizilemedi. Tek sınıf olabilir.")
        return

    plt.figure(figsize=(7, 5), facecolor="white")
    ax = plt.gca()
    ax.set_facecolor("white")

    plt.plot(fpr, tpr, color=BLUE, lw=2, label=f"ROC Curve - AUC = {auc_score:.3f}")
    plt.plot([0, 1], [0, 1], color=LIGHT_BLUE, lw=2, linestyle="--", label="Random Classifier")

    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title(f"ViT-L/32 Baseline - Fold {fold} - {split_name} ROC Curve")
    plt.legend(loc="lower right")
    plt.grid(True, alpha=0.3)

    save_and_show_plot(f"vit_l32_baseline_fold_{fold}_{split_name.lower()}_roc_curve.png")


# ============================================================
# 14. 5-FOLD CROSS VALIDATION
# ============================================================

targets = np.array([label for _, label in full_train_eval_dataset.samples])

skf = StratifiedKFold(
    n_splits=K_FOLDS,
    shuffle=True,
    random_state=42
)

fold_train_results = []
fold_validation_results = []
fold_test_results = []

print("\n--- ViT-L/32 Baseline 5-Fold Cross Validation Başlıyor ---")

for fold, (train_idx, val_idx) in enumerate(skf.split(np.zeros(len(targets)), targets), 1):

    print("\n" + "=" * 80)
    print(f"ViT-L/32 BASELINE FOLD {fold}/{K_FOLDS}")
    print("=" * 80)

    train_subset = Subset(full_train_aug_dataset, train_idx)
    train_eval_subset = Subset(full_train_eval_dataset, train_idx)
    val_subset = Subset(full_train_eval_dataset, val_idx)

    train_loader = DataLoader(train_subset, batch_size=BATCH_SIZE, shuffle=True, num_workers=2, pin_memory=torch.cuda.is_available())
    train_eval_loader = DataLoader(train_eval_subset, batch_size=BATCH_SIZE, shuffle=False, num_workers=2, pin_memory=torch.cuda.is_available())
    val_loader = DataLoader(val_subset, batch_size=BATCH_SIZE, shuffle=False, num_workers=2, pin_memory=torch.cuda.is_available())

    model = build_vit_l32_baseline_model()

    criterion = nn.CrossEntropyLoss()

    optimizer = optim.Adam(
        filter(lambda p: p.requires_grad, model.parameters()),
        lr=1e-4
    )

    history = {
        "train_loss": [],
        "train_acc": [],
        "val_loss": [],
        "val_acc": []
    }

    for epoch in range(NUM_EPOCHS):
        train_loss, train_acc = train_one_epoch(model, train_loader, optimizer, criterion)
        val_loss, val_acc = evaluate(model, val_loader, criterion)

        history["train_loss"].append(train_loss)
        history["train_acc"].append(train_acc)
        history["val_loss"].append(val_loss)
        history["val_acc"].append(val_acc)

        print(
            f"Fold {fold} | Epoch {epoch+1}/{NUM_EPOCHS} | "
            f"Train Loss: {train_loss:.4f} | Train Acc: {train_acc:.4f} | "
            f"Val Loss: {val_loss:.4f} | Val Acc: {val_acc:.4f}"
        )

    plot_accuracy(history, fold)
    plot_loss(history, fold)

    train_labels, train_preds, train_probs = get_predictions(model, train_eval_loader)
    train_metrics = calculate_metrics(train_labels, train_preds, train_probs)

    print(f"\n--- FOLD {fold} TRAIN SETİ SONUÇLARI ---")
    for metric_name, metric_value in train_metrics.items():
        print(f"{metric_name}: {metric_value:.4f}")

    plot_confusion_matrix(
        train_labels,
        train_preds,
        f"ViT-L/32 Baseline - Fold {fold} - Train Confusion Matrix",
        f"vit_l32_baseline_fold_{fold}_train_confusion_matrix.png"
    )

    plot_roc_curve(train_labels, train_probs, train_metrics["ROC-AUC"], fold, "Train")

    fold_train_results.append({
        "Fold": fold,
        "Final Train Loss": history["train_loss"][-1],
        "Final Train Accuracy": history["train_acc"][-1],
        **train_metrics
    })

    val_labels, val_preds, val_probs = get_predictions(model, val_loader)
    val_metrics = calculate_metrics(val_labels, val_preds, val_probs)

    print(f"\n--- FOLD {fold} VALIDATION SETİ SONUÇLARI ---")
    for metric_name, metric_value in val_metrics.items():
        print(f"{metric_name}: {metric_value:.4f}")

    plot_confusion_matrix(
        val_labels,
        val_preds,
        f"ViT-L/32 Baseline - Fold {fold} - Validation Confusion Matrix",
        f"vit_l32_baseline_fold_{fold}_validation_confusion_matrix.png"
    )

    plot_roc_curve(val_labels, val_probs, val_metrics["ROC-AUC"], fold, "Validation")

    fold_validation_results.append({
        "Fold": fold,
        "Final Validation Loss": history["val_loss"][-1],
        "Final Validation Accuracy": history["val_acc"][-1],
        **val_metrics
    })

    test_labels, test_preds, test_probs = get_predictions(model, test_loader)
    test_metrics = calculate_metrics(test_labels, test_preds, test_probs)

    print(f"\n--- FOLD {fold} TEST SETİ SONUÇLARI ---")
    for metric_name, metric_value in test_metrics.items():
        print(f"{metric_name}: {metric_value:.4f}")

    fold_test_results.append({
        "Fold": fold,
        **test_metrics
    })

    plot_test_metrics(test_metrics, fold)

    plot_confusion_matrix(
        test_labels,
        test_preds,
        f"ViT-L/32 Baseline - Fold {fold} - Test Confusion Matrix",
        f"vit_l32_baseline_fold_{fold}_test_confusion_matrix.png"
    )

    plot_roc_curve(test_labels, test_probs, test_metrics["ROC-AUC"], fold, "Test")

    torch.save(
        model.state_dict(),
        os.path.join(OUTPUT_DIR, f"vit_l32_baseline_fold_{fold}.pth")
    )


# ============================================================
# 15. TÜM FOLD SONUÇLARI
# ============================================================

train_results_df = pd.DataFrame(fold_train_results)
validation_results_df = pd.DataFrame(fold_validation_results)
test_results_df = pd.DataFrame(fold_test_results)

print("\n" + "=" * 80)
print("ViT-L/32 BASELINE TÜM FOLD TRAIN SONUÇLARI")
print("=" * 80)
print(train_results_df)

print("\n" + "=" * 80)
print("ViT-L/32 BASELINE TÜM FOLD VALIDATION SONUÇLARI")
print("=" * 80)
print(validation_results_df)

print("\n" + "=" * 80)
print("ViT-L/32 BASELINE TÜM FOLD TEST SONUÇLARI")
print("=" * 80)
print(test_results_df)

print("\n" + "=" * 80)
print("ViT-L/32 BASELINE TEST ORTALAMA ± STD")
print("=" * 80)

for col in ["Accuracy", "Precision", "Recall", "F1-Score", "ROC-AUC"]:
    mean_val = test_results_df[col].mean()
    std_val = test_results_df[col].std()
    print(f"{col}: {mean_val:.4f} ± {std_val:.4f}")

train_results_df.to_csv(os.path.join(OUTPUT_DIR, "vit_l32_baseline_train_results.csv"), index=False)
validation_results_df.to_csv(os.path.join(OUTPUT_DIR, "vit_l32_baseline_validation_results.csv"), index=False)
test_results_df.to_csv(os.path.join(OUTPUT_DIR, "vit_l32_baseline_test_results.csv"), index=False)

print(f"\nÇıktılar kaydedildi: {OUTPUT_DIR}")
```
